```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from matplotlib.pyplot import figure
```

#### Use Pandas to import the data from epa-sea-level.csv

```python
df = pd.read_csv('epa-sea-level.csv')
df.info()
```

```python
df.head()
```

#### Question: Use matplotlib to create a scatter plot using the "Year" column as the x-axis and the "CSIRO Adjusted Sea Level" column as the y-axix.

```python
figure(figsize=(14, 6), dpi=80)
plt.scatter(x='Year', y='CSIRO Adjusted Sea Level', data=df)
plt.xlabel('Year')
plt.ylabel('CSIRO Adjusted Sea Level')
```

#### Question: Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.

```python
#linregress function return model intercept and slope parameters
result = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
print(result.intercept, result.slope)
```

```python
#model parameters
intercept = result.intercept
slope = result.slope
```

```python
#use model paramteres to predict values upto the year 2050
x2 = np.arange(1850, 2051, 1)
y2 = []
for year in x2:
    y2.append(intercept + slope*year)
```

```python
#plot
figure(figsize=(14, 6), dpi=80)
plt.scatter(df.Year, df['CSIRO Adjusted Sea Level'], color = 'green')
plt.plot(x2, y2, 'r', label = 'Bests Fit Line 1')
plt.xlabel('Year')
plt.ylabel('Sea level')
```

#### Question: Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000. The x label should be "Year", the y label should be "Sea Level (inches)", and the title should be "Rise in Sea Level".

```python
#get data where year >= 2000
x = df.Year.loc[df['Year'] >= 2000]
y = df['CSIRO Adjusted Sea Level'].loc[df['Year'] >= 2000]
```

```python
#Input and output varaibe for libear regression model
x = np.array(x)
y = np.array(y)
```

```python
#linregress function return model intercept and slope parameters
result1 = linregress(x, y)
```

```python
#model parameters
intercept1 = result1.intercept
slope1 = result1.slope
```

```python
#use model paramteres to predict values upto the year 2050
x2 = np.arange(2000, 2051, 1)
y2 = []
for year in x2:
    y2.append(intercept1 + slope1*year)
```

```python
#plot
figure(figsize=(14, 6), dpi=80)
plt.scatter(x, y, color = 'green')
plt.plot(x2, y2, 'r')
plt.title('Sea Level Rise')
plt.xlabel('Year')
plt.ylabel('Sea level (inches)')
```
