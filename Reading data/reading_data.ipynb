```python
# using python to open a file using a pointer
with open ('btc-market-price.csv', 'r') as dt:
    print(dt)
```

```python
# read and print content
with open ('btc-market-price.csv', 'r') as dt:
    for index, line in enumerate(dt.readlines()):
        if (index<10):
            print(index, line)
```

```python
#using pandas
import pandas as pd
df = pd.read_csv('btc-market-price.csv')
df
```

```python
#formating it
df = pd.read_csv('btc-market-price.csv', header=None, parse_dates=True)
df
```

```python
#formating null values
df = pd.read_csv('btc-market-price.csv', header=None, parse_dates=True, na_values=['?', '-', ''])
df
```

```python
#setting column names
df = pd.read_csv('btc-market-price.csv', header=None, parse_dates=True, na_values=['?', '-', ''], names=['Time', 'Price'])
df
```

```python
#set datatype
df = pd.read_csv('btc-market-price.csv', header=None, parse_dates=True, 
                 na_values=['?', '-', ''], names=['Time', 'Price'], dtype={'Price' : 'float'})
df
```

```python
#changing time to a dattime datatype
df['Time'] = pd.to_datetime(df['Time'])
```
