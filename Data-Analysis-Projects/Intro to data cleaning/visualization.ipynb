```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
df = pd.read_csv('btc-eth-prices-outliers.csv', index_col=0, parse_dates=True)
df.info()
df.head()
```

```python
#simple visualization
df.plot(figsize=(16,10))
```

This plot shows the timeseries data clearly. However, there are some gaps in the plot. Missing values maybe?

```python
df.loc['2017-12-01' : '2017-12-15'].plot(y='Ether', figsize=(14,7))
```

There are three missing values. This can be fixed by bfill and ffill but I'm unsure. I don't think mean is a good estimate here because there is a rapid hike around the 11th of Dec. 

```python
#verifying null values
df.Ether.isnull().sum()
```

```python
#extract where the null values are
df.loc[df['Ether'].isnull()]
```

```python
#use backward fill
df.loc['2017-12-08' : '2012-12-10'].fillna(method='bfill')
df.Ether.isnull().sum()
#this didnt work because, like most functions, fillna() is immutable. 
```

```python
#use inplace
df.fillna(method='bfill', inplace=True)
```

```python
df.isnull().sum()
```

```python
#check plot
df.plot(figsize=(16,10))
```

Usually, one must avoid removing data However, the two spikes in both ETH and Bitcoin seem like extreme (very) outliers. These can be removed in my opinion. Let's first visualize them

```python
#first outlier peak
df.loc['2017-12-20': '2018-01-02'].plot()
```

```python
#second outlier peak
df.loc['2018-03-01': '2018-03-15'].plot()
```

```python
#drop these two values and create a new df1
df1 = df.drop(['2018-03-04', '2017-12-28'])
```
